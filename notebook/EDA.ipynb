```python
#Filling Outliers

def Capping_IQR(data,col,factor):
    for i in col:
        q1=data[i].quantile(0.25)
        q3=data[i].quantile(0.75)
        iqr=q3-q1
        upper=q3+(iqr*factor)
        lower=q1-(iqr*factor)
        data[i]=np.where(data[i]>upper,upper,np.where(data[i]<lower,lower,data[i]))
```

```python
Capping_IQR(data,numeric_features_outlier,1.5)
```
